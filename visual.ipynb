```python
import matplotlib.pyplot as plt
import matplotlib.tri as tri
import numpy as np
from collections import defaultdict
from matplotlib.patches import Polygon



partitions = defaultdict(list)
```

```python



with open("partitions.txt", "r") as f:
    for line in f:
        *coords, part = line.strip().split()
        coords = list(map(float, coords))
        part = int(part)
        #Each triangle = 3 vertices = 6 floats
        triangle = np.array([[coords[0], coords[1]],
                             [coords[2], coords[3]],
                             [coords[4], coords[5]]])
        partitions[part].append(triangle)


plt.figure(figsize=(8, 8))
colors = plt.cm.tab20.colors

for i, tris in partitions.items():
    for t in tris:
        tri_plot = plt.Polygon(t, edgecolor='black', facecolor=colors[i % len(colors)], linewidth=0.5)
        plt.gca().add_patch(tri_plot)

plt.gca().set_aspect('equal')
plt.title("2D Triangular Mesh Colored by Partition")
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout()
plt.show()
```
